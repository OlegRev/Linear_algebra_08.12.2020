### __5.__ –£—Å—Ç–∞–Ω–æ–≤–∏—Ç—å, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –ª–∏–Ω–µ–π–Ω—ã–º –ø–æ–¥–ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ–º:

–±) –≤—Å–µ –≤–µ–∫—Ç–æ—Ä—ã, —è–≤–ª—è—é—â–∏–µ—Å—è –ª–∏–Ω–µ–π–Ω—ã–º–∏ –∫–æ–º–±–∏–Ω–∞—Ü–∏—è–º–∏ –¥–∞–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ $\{u_{1}, u_{2}, ..., u_{n}\}$.

## –ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–æ–µ –∑–∞–¥–∞–Ω–∏–µ

### __1.__ –ù–∞–π—Ç–∏ —Å–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–æ–≤ $x, y \in \mathbb{R}$:<br>
–∞) $x=(0,-3, 6),~y=(-4, 7, 9);$<br>
–±) $x=(7, -4, 0, 1),~y=(-3, 1, 11, 2).$


```python
xa = [0, -3, 6]
ya = [-4, 7, 9]
a = xa[0]*ya[0]+xa[1]*ya[1]+xa[2]*ya[2]
xb = [7, -4, 0, 1]
yb = [-3, 1, 11, 2]
b = xb[0]*yb[0] + xb[1]*yb[1] + xb[2]*yb[2] + xb[3]*yb[3]


xa1 = np.array(xa)
ya1 = np.array(ya)
xb1 = np.array(xb)
yb1 = np.array(yb)

print(f'–°–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –ø–æ—Å—á–∏—Ç–∞–Ω–æ–µ –≤ —Ä—É—á–Ω—É—é:\n\t–∞): {a}\n\t–±):{b}\n\n'
      f'–°–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ numpy.dot(x, y) a) ùë•=(0,‚àí3,6), ùë¶=(‚àí4,7,9):\n'
      f'\t{np.dot(xa1, ya1)}\n'
      f'–°–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ numpy.dot(x, y) –±) ùë•=(7,‚àí4,0,1), ùë¶=(‚àí3,1,11,2):\n'
      f'\t{np.dot(xb1, yb1)}')
```

### __2.__ –ù–∞–π—Ç–∏ –Ω–æ—Ä–º—ã –≤–µ–∫—Ç–æ—Ä–æ–≤ $(4, 2, 4)$ –∏ $(12, 3, 4)$ –∏ —É–≥–æ–ª –º–µ–∂–¥—É –Ω–∏–º–∏.

```python
x = [4,2,4]
y = [12,3,4]

l1x = abs(x[0]) + abs(x[1]) + abs(x[2])
l1y = abs(y[0]) + abs(y[1]) + abs(y[2])

l2x = np.sqrt(x[0]**2 + x[1]**2 + x[2]**2)
l2y = np.sqrt(y[0]**2 + y[1]**2 + y[2]**2)
cos_f = (x[0]*y[0]+x[1]*y[1]+x[2]*y[2])/(l2x*l2y)
x1=np.array([4,2,4])
y1=np.array([12,3,4])
print(f'–ù–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ (4,2,4) l1 ={l1x}\n'
      f'–ù–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ (12,3,4) l1 ={l1y}\n'
      f'–ù–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ (4,2,4) l2 ={l2x}\n'
      f'–ù–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ (12,3,4) l2 ={l2y}\n'
      f'cos_f –º–µ–∂–¥—É –≤–µ–∫—Ç–æ—Ä–∞–º–∏  (4,2,4) –∏ (12,3,4) : {cos_f}\n'
      f'—É–≥–æ–ª –º–µ–∂–¥—É –≤–µ–∫—Ç–æ—Ä–∞–º–∏ —Ä–∞–≤–µ–Ω {np.arccos(cos_f)}')
print(f'l1 –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞  (4,2,4):\n{norm(x1, ord=1)}\n'
      f'l1 –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ (12,3,4) :\n{norm(y1, ord=1)}\n'
      f'l2 –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞  (4,2,4):\n{norm(x1, ord=2)}\n'
      f'l2 –Ω–æ—Ä–º–∞ –≤–µ–∫—Ç–æ—Ä–∞ (12,3,4) :\n{norm(y1, ord=2)}')
```

### __3.__ –ë—É–¥–µ—Ç –ª–∏ –ª–∏–Ω–µ–π–Ω–æ–µ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ –µ–≤–∫–ª–∏–¥–æ–≤—ã–º, –µ—Å–ª–∏ –∑–∞ —Å–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –ø—Ä–∏–Ω—è—Ç—å:
    –∞) –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –¥–ª–∏–Ω –≤–µ–∫—Ç–æ—Ä–æ–≤;
    –±) —É—Ç—Ä–æ–µ–Ω–Ω–æ–µ –æ–±—ã—á–Ω–æ–µ —Å–∫–∞–ª—è—Ä–Ω–æ–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–æ–≤?
   
   


__4.__ –ö–∞–∫–∏–µ –∏–∑ –Ω–∏–∂–µ–ø–µ—Ä–µ—á–∏—Å–ª–µ–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ –æ–±—Ä–∞–∑—É—é—Ç –æ—Ä—Ç–æ–Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –±–∞–∑–∏—Å –≤ –ª–∏–Ω–µ–π–Ω–æ–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ $\mathbb{R}^{3}$:<br>
–∞) $(1,0,0),(0,0,1);$ <br>
–±) $(1/\sqrt{2},-1/\sqrt{2},0),(1/\sqrt{2},1/\sqrt{2},0), (0,0,1);$<br>
–≤) $(1/2, -1/2, 0), (0, 1/2, 1/2), (0,0,1);$<br>
–≥) $(1,0,0),(0,1,0),(0,0,1)?$ 

–≥)

```python
a = np.array([1/(2)**0.5,-1/(2)**0.5,0])
b = np.array([1/(2)**0.5,1/(2)**0.5,0])
c = np.array([0,0,1])

print(np.dot(a, a),np.dot(b, b),np.dot(c, c))
print(f'(a,b){np.dot(a,b)}\t(a,c){np.dot(a, c)}\t(c, b){np.dot(c, b)}')
```
